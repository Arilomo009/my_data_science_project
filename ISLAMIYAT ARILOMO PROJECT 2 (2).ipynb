1. Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load Dataset

```python
cases = pd.read_csv("telco_customer_churn2.csv")
cases.head()
```

2. Check Data Structure

2a. Check shapes

```python
cases.shape
```

2b. Check Column

```python
cases.columns
```

3. Clean column

```python
cols = ["customer_id","gender","senior_citizen","partner","dependents","tenure",
        "phone_service","multiple_lines","internet_service","online_security",
        "online_backup","device_protection","tech_support","streaming_tv",
        "streaming_movies","contract","paperless_billing","payment_method",
        "monthly_charges","total_charges","churn"]

cases.columns = cols

cases.head()
```

4. Check data types

```python
cases.info()
```

5. Summary statistics

```python
cases.describe()
```

6. Checking missing values

```python
cases.isnull().sum()
```

7. Removing rows that contains missing values

```python
cases = cases.dropna()
cases.isnull().sum
```

8. VISUALISATION-1

```python
sns.countplot(x="churn", data=cases)

plt.title("Distribution of Customer Churn")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

INTERPRETATION_1

The chart illustrates the distribution of customers who remained with the telecommunications service compared to those who discontinued it. This visualization facilitates an assessment of the prevalence of customer churn within the dataset. A high frequency of affirmative responses ("Yes") indicates that customer retention could represent a significant challenge for the company.

VISUALISATION-2

```python
sns.countplot(x="contract", hue="churn", data=cases)

plt.title("Customer Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.show()
```

INTERPRETATION_2

The chart presents a comparative analysis of churn rates among various contract types. It is observed that customers engaged in month-to-month contracts exhibit higher churn rates compared to those bound by long term agreements. This finding implies that extended contractual commitments may contribute to a reduction in customer attrition.

VISUALISATION-3

```python
sns.boxplot(x="churn", y="monthly_charges", data=cases)

plt.title("Monthly Charges and Customer Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

INTERPRETATION-3

The boxplot illustrates the distribution of monthly charges among customers who discontinued their service compared to those who remained. A tendency for higher charges among customers who churned may indicate that price sensitivity plays a role in customer attrition.

9. FINAL REFLECTION

The analysis of customer churn within the telecommunications dataset uncovers several significant patterns. Visual representations indicate a substantial proportion of customers discontinue the service, underscoring the necessity of identifying the determinants of churn. Contract type emerges as a key factor, with customers engaged in month-to-month agreements exhibiting a higher propensity to churn compared to those bound by longer-term contracts. This finding implies that extended contract durations may contribute to enhanced customer retention. Furthermore, the observed correlation between monthly charges and churn suggests that customers incurring higher monthly fees are more inclined to terminate the service, potentially due to financial considerations. The distribution of customer tenure reveals that individuals with shorter durations of service are more prone to churn, highlighting the critical importance of the initial customer experience. Collectively, these insights indicate that telecommunications providers should prioritize optimizing onboarding processes, implementing competitive pricing strategies, and promoting longer-term contractual commitments to mitigate churn and foster customer loyalty.
